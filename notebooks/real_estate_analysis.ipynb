```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
import os
from datetime import datetime
```

```python
# Define the URL to scrape
url = 'https://www.redfin.com/city/29470/IL/Chicago'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the HTML content of the page
    html_content = response.text
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
# Parse the HTML content with BeautifulSoup
soup = bs(html_content, 'html.parser')
```

```python
# Extract the infromation from the soup with a class
house_prices = soup.find_all('span', {"class": 'homecardV2Price'})
house_adresses = soup.find_all('span', {"class": 'collapsedAddress primaryLine'})
house_stats = soup.find_all('div', {"class": 'stats'})
```

```python
# Create an empty dictionary to store the scraped data by extracting house 
house_obj = {
    "House_Prices": [price.text for price in house_prices],
    "House_Adress": [address.text for address in house_adresses],
    "Num_of_Beds": [house_stats[i].text for i in range(0, len(house_stats), 3)],
    "Num_of_Baths": [house_stats[i + 1].text for i in range(0, len(house_stats), 3)],
    "Square_Footage": [house_stats[i+2].text.split(" ")[0] for i in range(0, len(house_stats), 3)],
}

# Create a Pandas DataFrame from the scraped data
house_dataset = pd.DataFrame(house_obj)

# Display the resulting DataFrame
house_dataset
```

```python
import pandas as pd
import os
from datetime import datetime

# Define the path to the 'data' folder within your project directory
data_folder = "../data"

# Ensure the 'data' folder exists; if not, create it
if not os.path.exists(data_folder):
    os.makedirs(data_folder)

# Generate a timestamp for the filename
timestamp = datetime.now().strftime("%Y.%m.%d")

# Create a unique filename for the CSV file in the 'data' folder using the timestamp
csv_filename = os.path.join(data_folder, f"home_data_{timestamp}.csv")

# Save the DataFrame to the CSV file
house_dataset.to_csv(csv_filename, index=False)

# Print the generated filename
print(f"Data saved to {csv_filename}")
```
